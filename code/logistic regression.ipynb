```python
def sigmoid(z):
    return 1.0/(1+np.exp(-z))
```

```python
def cost(theta,X,y):
    first = np.multiply(y,np.log(sigmoid(X*theta.T)))
    second = np.multiply((1-y),np.log(1-sigmoid(X*theta.T)))
    return -np.sum(first + second)/len(y)
```

```python
data.insert(0,'ones',1)
theta = np.zeros(3)
cols = data.shape[1]
X = data.iloc[:,0:cols-1]
y = data.iloc[:,cols-1:cols]
```

```python
X = np.mat(X.values)
y = np.mat(y.values)
```

```python
def train(theta,X,y,alpha,maxIterations):
    for k in range(maxIterations):
        for i in range(len(y)):
            for j in range(theta.shape[1]):
                theta[0,j] -= alpha*(((sigmoid(X[i,:]*theta.T) - y[i,0]) * X[i,j]))
    costvalue = cost(theta,X,y)
    return theta,costvalue
```

```python
alpha = 0.03
maxIterations = 3500
par,costvalue = train(theta,X,y,alpha,maxIterations)
```

```python
def predict(theta,X):
    probability = sigmoid(X*theta.T)
    return [1 if i>=0.5 else 0 for i in probability]
```

```python
predictions = predict(par,X)
correct = [1 if (a==1 and b==1)or(a==0 and b==0) else 0 for (a,b) in zip(predictions,y)];
accuracy = float(np.sum(correct))/len(y)
print('accuracy is {}%'.format(accuracy*100))
```
